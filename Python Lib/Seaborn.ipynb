```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
sns.set_theme()
```

```python
plt.figure(figsize=(8,5))
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue='smoker',size='size');
```

```python
sns.relplot(tips, x='total_bill', y='tip', hue='day')
```

```python
plt.figure(figsize = (10,5))
sns.relplot(tips, x='total_bill', y='tip', hue='time', col='day', row='smoker')
```

Columns can be wrapped across multiple rows.

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", col="day", col_wrap=2)
plt.show()
```

# Distributional Representation

```python
tips.head()
```

```python
sns.displot(tips, x='total_bill', col='time', kde = True)
```

Calculating and plotting empirical cumulative distribution function

```python
sns.displot(tips, kind='ecdf', x='total_bill', col='time', hue = 'smoker', rug=True)
```

```python
#showing mean value in the bar graph where ever we have duplucation based on visual criteria
sns.barplot(tips, x='day',y='total_bill', hue='smoker')
```

```python
sns.catplot(tips, x='day',y='total_bill',hue='smoker', kind='bar')
```

```python
sns.catplot(tips, x='day',y='total_bill',hue='smoker', kind='swarm')
```

```python
sns.catplot(tips, x='day',y='total_bill',hue='smoker', kind='violin')
```

```python
#using split function with hue to show 2 graphs side by side
sns.catplot(tips, x='day',y='total_bill',hue='smoker', kind='violin', split=True)
```
